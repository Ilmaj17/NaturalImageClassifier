```python
X_data = []
y_data = []
for l in label:
    path = '/kaggle/input/natural-images/natural_images/'+l
    folder = os.listdir(path)
    #print(path)
    
    for f in folder:
        path2 = path + '/'+f
        #print(path2)
        img = cv2.imread(path2)
        resized_img = cv2.resize(img, (256, 256))
        img_arr = np.array(resized_img)
        X_data.append(img_arr)
        y_data.append(l)
```
